```python
import pandas as pd
```

```python
df0 = pd.read_csv('creditcard.csv')
df0.head()
```

```python
df0.isnull().sum()
```

```python
describe = df0.describe()
describe
```

```python
class_count = df0.Class.value_counts()
class_count
```

```python
df1 = df0[df0.Class == 0]
df1 = df1[::len(df1) // class_count[1]]
df1.head()
```

```python
df2 = pd.concat([df1, df0[df0.Class == 1]])
df2.head()
```

```python
df2.Class.value_counts()
```

```python
x = df2[df2.columns[1:-2]]
y = df2[df2.columns[-1]]
x.head()
```

```python
scale = describe[describe.columns[1:-2]]
scale
```

```python
x_scale = (x - scale.loc['mean']) / scale.loc['std']
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.2, random_state=0)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=0)
print('Train size =', len(y_train))
print('Validation size =', len(y_val))
print('Test size =', len(y_test))
x_train.head()
```

```python
svm_c = [100 * 0.5 ** i for i in range(10)]
svm_c[:5]
```

```python
from sklearn.svm import SVC
from sklearn.metrics import f1_score
svm_models = []
svm_scores = []
for c in svm_c:
    svm_models.append(SVC(C=c).fit(x_train, y_train))
    svm_scores.append(f1_score(y_val, svm_models[-1].predict(x_val), average=None))
svm_best_score = 0
svm_best_score_index = -1
for i in range(len(svm_scores)):
    if svm_best_score < sum(svm_scores[i]) / len(svm_scores[i]):
        svm_best_score = sum(svm_scores[i]) / len(svm_scores[i])
        svm_best_score_index = i
print('Best validation score =', svm_scores[svm_best_score_index])
svm_test_score = f1_score(y_test, svm_models[svm_best_score_index].predict(x_test), average=None)
svm_test_score
```

```python
mlp_alpha = [0.0001 * 2 ** i for i in range(10)]
mlp_alpha[:5]
```

```python
from sklearn.neural_network import MLPClassifier
mlp_models = []
mlp_scores = []
for a in mlp_alpha:
    mlp_models.append(MLPClassifier(alpha=a, max_iter=500).fit(x_train, y_train))
    mlp_scores.append(f1_score(y_val, mlp_models[-1].predict(x_val), average=None))
mlp_best_score = 0
mlp_best_score_index = -1
for i in range(len(mlp_scores)):
    if mlp_best_score < sum(mlp_scores[i]) / len(mlp_scores[i]):
        mlp_best_score = sum(mlp_scores[i]) / len(mlp_scores[i])
        mlp_best_score_index = i
print('Best validation score =', mlp_scores[mlp_best_score_index])
mlp_test_score = f1_score(y_test, mlp_models[mlp_best_score_index].predict(x_test), average=None)
mlp_test_score
```

```python
model = svm_models[svm_best_score_index] if sum(svm_test_score) / len(svm_test_score) > sum(mlp_test_score) / len(mlp_test_score) else mlp_models[mlp_best_score_index]
submission = pd.DataFrame({'Class': model.predict(x_test)})
submission.head()
```

```python
submission.to_csv('submission.csv', index=False)
```
